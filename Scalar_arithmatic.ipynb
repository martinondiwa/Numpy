```python
#Scalar arithmatic operations
#Scaler is arithmatic functions that means a single value
```

```python
import numpy as np
```

```python
#we need an array to work with
array = np.array([1,2,3])

#adding value to each element of the array
print(array + 1)

#subtracting value from each element of the array
print(array - 1)

#multiplication
print(array * 3)

#division
print(array / 2)

#raising each element to the power of 5
print(array ** 5)

#
print(array//3)
```
