```python
#Creating a numpy array
array = np.array([1, 2, 3, 4])
print(array)
```

```python
#multidimensional array
array = np.array([[1, 2, 3], [4, 5, 6]])
print(array)
print(array.ndim)

```

```python
#when you arrange them downwards they form like a 2D matrix
#to achieve 2d array close the matrix within another matrix and add another list
array = np.array([['k', 'g', 'f', 'h'],
                  ['a', 'b', 'c', 'd'],
                  ['g', 'g', 'f', 'r']])
print(array.ndim)
```

```python
#3D array
```

```python
#when you arrange them downwards they form like a 2D matrix
#pay attention to the number of brackets
array = np.array([[[1, 2, 3, 4]],
                  [['a', 'b', 'c', 'd']],
                  [['g', 'g', 'f', 'r']]])

print(array.ndim)
print (array.shape)
```

```python
#when you arrange them downwards they form like a 2D matrix
#pay attention to the number of brackets
array = np.array([[[1, 2, 3, 4], ],
                  [['a', 'b', 'c', 'd']],
                  [['g', 'g', 'f', 'r']]])

print(array.ndim)
print (array.shape)
```

```python
#when you arrange them downwards they form like a 2D matrix
#pay attention to the number of brackets
array = np.array([[[1, 2, 3, 4], [1, 2, 3, 4]],
                  [['a', 'b', 'c', 'd'], ['s', 'e', 't,', 'z']],
                  [['g', 'g', 'f', 'r'], ['t', 'f', 'g', 'p']]])

print(array.ndim)
print (array.shape)

# to get elements in our array we shall use their indices eaxample
#getting the first element
print(array[0][0][0])  #this is called chain indexing

#Their is also multidimensional indexing
print(array[0, 0, 0])
print(array[1, 0, 2])
print(array[1, 1, 1])

#Using string concatination to print 3 letter word from the multidimensional array
word = array[1, 0, 1] + array[0, 0, 1] + array[0, 0, 3]
print(word)
```
