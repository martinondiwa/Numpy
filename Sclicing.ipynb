```python
#column selection
array = np.array([[1, 2, 3, 4],
                 [5, 6, 7, 8],
                 [13, 14, 15, 16],
                 [9, 10, 11, 12]])
#for columns you must have something on first value ':'
print(array[:, 0])
print(array[:, 0:3]) #prints first 3 columns
print(array[:, 0:3:2]) #prints first 3 columns with a step of 2
print(array[:, -1]) #prints last column
print(array[:, :: 2]) #prints every second column

#printing a row and a column
print(array[0:2, 0:2]) #gives the first two columns of the first two rows
#print(array[0, 0:3])
print(array[2:, 0:2])
#last two rows last two columns
print(array[2:, 2:])
```

```python
array = np.array([[1, 2, 3, 4],
                 [5, 6, 7, 8],
                 [13, 14, 15, 16],
                 [9, 10, 11, 12]])
#array[start:end:step]
print(array[0])
#selecting a range
print(array[0:3])
print(array[0:3:2])
#Negative indexing
print(array[-1])
print(array[-1:-3])
print(array[-1:-3:-1])
#printing all the rows reversed
print(array[::-1])
```
