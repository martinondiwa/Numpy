```python
import numpy as np
```

```python
array = np.array([[1, 2, 3, 4],
                 [5, 6, 7, 8],
                 [13, 14, 15, 16],
                 [9, 10, 11, 12]])
#array[start:end:step]
print(array[0])
#selecting a range
print(array[0:3])
print(array[0:3:2])
#Negative indexing
print(array[-1])
print(array[-1:-3])
print(array[-1:-3:-1])
#printing all the rows reversed
print(array[::-1])
```
