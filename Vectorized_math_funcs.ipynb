```python
import numpy as np
```

```python
# vectorized math funcs
```

```python
array = np.array([1, 2, 3])

#getting the squaroot of the elements
print(np.sqrt(array))

#getting the exponential of the elements
print(np.exp(array))
```

```python
array = np.array([2.34, 1.45, 7.7])

#rounding off
print(np.round(array))
```

```python
#rounding down
array = np.array([1.2, 3.5, 6.86])


print(np.floor(array))
```
